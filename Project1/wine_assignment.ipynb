```python
import pandas as pd
```

```python
df=pd.read_csv('wines.csv')
```

```python
df['new_quality']=[1 if x>=7 else 0 for x in df['quality']]
```

```python
x=df.iloc[:,0:-2]
```

```python
x
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
x_scaled
```

```python
X=x_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=42)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier(random_state=1)
```

```python
dtc.fit(X_train,y_train)
```

```python
dtrpred=dtc.predict(X_test)
```

```python
print(accuracy_score(dtrpred,y_test))
print(confusion_matrix(dtrpred,y_test))
print(classification_report(dtrpred,y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc=GradientBoostingClassifier(random_state=1)
```

```python
gbc.fit(X_train,y_train)
gbcpred=gbc.predict(X_test)

```

```python
print(accuracy_score(gbcpred,y_test))
print(confusion_matrix(gbcpred,y_test))
print(classification_report(gbcpred,y_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
abc=AdaBoostClassifier(random_state=1)
```

```python
abc.fit(X_train,y_train)
abcpred=abc.predict(X_test)
```

```python
print(accuracy_score(abcpred,y_test))
print(confusion_matrix(abcpred,y_test))
print(classification_report(abcpred,y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
gnb.fit(X_train,y_train)
gnbpred=gnb.predict(X_test)
print(accuracy_score(gnbpred,y_test))
print(confusion_matrix(gnbpred,y_test))
print(classification_report(gnbpred,y_test))
```

```python
import matplotlib.pyplot as plt
```

```python
important=pd.Series(gbc.feature_importances_,index=x.columns)
```

```python
important.nlargest(25).plot(kind='barh',figsize=(10,8))
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
y_pred_prob=dtc.predict_proba(X_test)[:,1]
```

```python
fpr,tpr,threshold=roc_curve(y_test,y_pred_prob)
```

```python
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='DecisionTreeClassifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('DecisionTreeClassifier')
plt.show()
```
