```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Grades.csv')
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df.shape
```

```python
x=df.iloc[:,1:-1]
```

```python
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
x.columns
```

```python
df.isnull().sum()
```

```python
for col in x.columns:
    x[f'{col}'].fillna(x[f'{col}'].mode()[0], inplace = True)
```

```python
x.isnull().sum()
```

```python
X = {'A+': 15, 'A': 14, 'A-': 13, 'B+': 12, 'B': 11, 'B-': 10,
    'C+': 9, 'C': 8, 'C-': 7, 'D+': 6, 'D': 5, 'F': 4,
     'I' : 3, 'W': 2, 'WU': 1}
```

```python
x = x.replace(X)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20,random_state=42)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lrpred=lr.predict(x_test)
```

```python
lrscore=lr.score(x_test,y_test)
```

```python
lrscore
```

```python
print("i - Mean Squared Error:", (mean_squared_error(lrpred,y_test)))
print("ii - Root Mean Squared Error:", np.sqrt(mean_squared_error(lrpred,y_test)))
print("iii - Mean Absolute Error:",(mean_absolute_error(lrpred,y_test)))
```

```python
ls=Lasso()
```

```python
ls
```

```python
ls.fit(x_train,y_train)
```

```python
ls.score(x_train,y_train)
```

```python
lspred=ls.predict(x_test)
```

```python
lrscore
```

```python
print("i - Mean Squared Error:", (mean_squared_error(lspred,y_test)))
print("ii - Root Mean Squared Error:", np.sqrt(mean_squared_error(lspred,y_test)))
print("iii - Mean Absolute Error:",(mean_absolute_error(lspred,y_test)))
```

```python
rd=Ridge()
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
rd.score
rdpred=rd.predict(x_test)

print("i - Mean Squared Error:", (mean_squared_error(rdpred,y_test)))
print("ii - Root Mean Squared Error:", np.sqrt(mean_squared_error(rdpred,y_test)))
print("iii - Mean Absolute Error:",(mean_absolute_error(rdpred,y_test)))
```

```python
el=ElasticNet()
el.fit(x_train,y_train)
el.score(x_train,y_train)
el.score
elpred=el.predict(x_test)

print("i - Mean Squared Error:", (mean_squared_error(elpred,y_test)))
print("ii - Root Mean Squared Error:", np.sqrt(mean_squared_error(elpred,y_test)))
print("iii - Mean Absolute Error:",(mean_absolute_error(elpred,y_test)))
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
dtr.score(x_train,y_train)
dtr.score
dtrpred=dtr.predict(x_test)

print("i - Mean Squared Error:", (mean_squared_error(dtrpred,y_test)))
print("ii - Root Mean Squared Error:", np.sqrt(mean_squared_error(dtrpred,y_test)))
print("iii - Mean Absolute Error:",(mean_absolute_error(dtrpred,y_test)))
```

```python
svr=SVR()
```

```python

svr.fit(x_train,y_train)
svr.score(x_train,y_train)
svr.score
svrpred=svr.predict(x_test)

print("i - Mean Squared Error:", (mean_squared_error(svrpred,y_test)))
print("ii - Root Mean Squared Error:", np.sqrt(mean_squared_error(svrpred,y_test)))
print("iii - Mean Absolute Error:",(mean_absolute_error(svrpred,y_test)))
```
